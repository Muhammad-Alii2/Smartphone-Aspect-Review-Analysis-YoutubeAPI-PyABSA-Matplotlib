```python
from googleapiclient.discovery import build
from youtube_transcript_api import YouTubeTranscriptApi
import textwrap
from pyabsa import ATEPCCheckpointManager
import matplotlib.pyplot as plt

def search_youtube(api_key, channel_id, keyword, max_results=1):
    # Create a YouTube Data API service
    youtube = build('youtube', 'v3', developerKey=api_key)

    # Search for videos from the specified channel and matching the keyword
    search_response = youtube.search().list(
        q=keyword,
        channelId=channel_id,
        type='video',
        part='id',
        maxResults=max_results
    ).execute()

    video_ids = [search_result['id']['videoId'] for search_result in search_response.get('items', [])]

    if not video_ids:
        print(f"No videos found for the keyword: {keyword}")
        return None

    video_id = video_ids[0]

    return video_id

def get_transcript_text(video_id):
    try:
        # Get the transcript for the video
        transcript = YouTubeTranscriptApi.get_transcript(video_id)

        # Extract text from the transcript
        transcript_text = ' '.join(entry['text'] for entry in transcript)

        return transcript_text
    except Exception as e:
        print(f"An error occurred while fetching the transcript: {e}")

    return None

def wrap_text_into_lines(text, max_words_per_line):
    # Split the text into words
    words = text.split()

    # Join words into lines with a maximum of max_words_per_line words per line
    data_to_process = textwrap.wrap(' '.join(words), width=max_words_per_line)

    return data_to_process

def extract_aspects_and_sentiments(inference_source, checkpoint='english', auto_device=True):
    # Get the aspect extractor using the specified checkpoint and device setting
    aspect_extractor = ATEPCCheckpointManager.get_aspect_extractor(checkpoint=checkpoint, auto_device=auto_device)

    # Extract aspects and sentiments
    aepc_result = aspect_extractor.extract_aspect(inference_source=inference_source,
                                                  pred_sentiment=True)

    return aepc_result
#
def extract_data_for_aspects(data_list, target_aspects):
    # sentiment_mapping = {'Positive': 1, 'Negative': -1}
    sentiment_mapping = {'Positive': 1, 'Neutral':0, 'Negative': -1}
    aspect_data = {aspect: {"sentiments": [], "confidences": []} for aspect in target_aspects}

    for data_entry in data_list:
        aspects = data_entry.get("aspect", [])
        sentiment = data_entry.get("sentiment", [])
        confidence = data_entry.get("confidence", [])

        for target_aspect in target_aspects:
            if any(target_aspect.lower() == aspect.lower() for aspect in aspects):
                target_index = next((i for i, a in enumerate(aspects) if a.lower() == target_aspect.lower()), None)

                if target_index is not None:
                    numeric_sentiment = sentiment_mapping.get(sentiment[target_index], 0)
                    aspect_data[target_aspect]["sentiments"].append(numeric_sentiment)
                    aspect_data[target_aspect]["confidences"].append(float(confidence[target_index]))

    return aspect_data

def calculate_rating_for_aspect(aspect_data):
    # Initialize lists to store confidences for each sentiment category
    positive_confidences = []
    negative_confidences = []
    neutral_confidences = []

    # Extract sentiments and confidences from the aspect_data dictionary
    sentiments = aspect_data.get("sentiments", [])
    confidences = aspect_data.get("confidences", [])

    if not sentiments or not confidences:
        print("No sentiments or confidences found. Unable to calculate rating.")
        return None

    # Iterate through each sentiment and confidence
    for sentiment, confidence in zip(sentiments, confidences):
        # Categorize the confidence based on sentiment value
        if sentiment == 1:
            positive_confidences.append(confidence)
        elif sentiment == -1:
            negative_confidences.append(confidence)
        else:
            neutral_confidences.append(confidence)

    # Calculate the average confidence for each sentiment category
    average_positive_confidence = sum(positive_confidences) / len(positive_confidences) if positive_confidences else 0.0
    average_negative_confidence = sum(negative_confidences) / len(negative_confidences) if negative_confidences else 0.0
    average_neutral_confidence = sum(neutral_confidences) / len(neutral_confidences) if neutral_confidences else 0.0

    # Print or use the average confidences for each sentiment category
    print("Average Confidence for Positive Sentiments:", average_positive_confidence)
    print("Average Confidence for Negative Sentiments:", average_negative_confidence)
    print("Average Confidence for Neutral Sentiments:", average_neutral_confidence)

    if not sentiments:
        print("No sentiments found. Unable to calculate rating.")
        return None

    # Calculate the unique sentiments and their counts
    unique_sentiments = set(sentiments)
    sentiment_counts = {sentiment: sentiments.count(sentiment) for sentiment in unique_sentiments}

    if not sentiment_counts:
        print("No sentiment counts found. Unable to calculate rating.")
        return None

    # Find the modes (most occurring sentiments)
    modes = [sentiment for sentiment, count in sentiment_counts.items() if count == max(sentiment_counts.values())]

    if not modes:
        print("No modes found. Unable to calculate rating.")
        return None

    # If there are multiple modes, choose the one with the highest average confidence
    chosen_sentiment = max(modes, key=lambda sentiment: sum(conf for s, conf in zip(sentiments, confidences) if s == sentiment) / sentiments.count(sentiment))
    mean_sentiment = chosen_sentiment
    # Print or use the chosen sentiment as the mean sentiment
    print("Mean Sentiment:", chosen_sentiment)

    # Assuming mean_sentiment, avg_pos_conf, avg_neg_conf, avg_neutral_conf are already calculated

    if mean_sentiment != 0 and average_positive_confidence != 0 and average_negative_confidence != 0:
        result = (average_positive_confidence - average_negative_confidence)
        result = abs(result)
        if average_neutral_confidence != 0:
            result = (result + average_neutral_confidence) / 2
    elif mean_sentiment != 0 and average_neutral_confidence != 0 and average_positive_confidence != 0:
        result = (average_positive_confidence + average_neutral_confidence) / 2
    elif mean_sentiment != 0 and average_neutral_confidence != 0 and average_negative_confidence != 0:
        result = (average_negative_confidence + average_neutral_confidence) / 2
    elif mean_sentiment == 0:
        # Assuming avg_conf_0 is already calculated
        result = (average_neutral_confidence * (20/100)) + 0.4
    else:
        result = sum(conf for s, conf in zip(sentiments, confidences) if s == chosen_sentiment) / sentiments.count(chosen_sentiment)

    if mean_sentiment == -1:
        result = 1 - result

    return result

def plot_rating_bar_chart(aspect_ratings,keyword):
    aspects = list(aspect_ratings.keys())
    ratings = [rating * 10.0 for rating in aspect_ratings.values()]
    bar_width = 0.2

    plt.figure(figsize=(len(aspect_ratings.keys())*1.5, 8))
    plt.bar(aspects, ratings,width=bar_width ,color='green')
    plt.xlabel('Aspects')
    plt.ylabel('Ratings (out of 10)')
    plt.title(f'Aspect Ratings for {keyword}')

    # Set y-axis limits to cover the full range from 0 to 10
    # plt.ylim(0, 10)
    plt.yticks(range(11))

    plt.show()

# Main usage
api_key = 'Your Youtube-API here'
# GSMArena, Marques Brownlee (MKBHD), Unbox Therapy
channel_ids = ['UCbLq9tsbo8peV22VxbDAfXA', 'UCBJycsmduvYEL83R_U4JriQ', 'UCsTcErHg8oDvUnTzoqsYeNw']
keyword = 'apple iphone 15 pro max full review'
target_aspects = ["IOS", "Android", "display", "battery", "chipset", "processor", "software", "network",
                  "design", "selfie Cam", "main Cam", "memory", "storage", "performance", "CPU", "GPU",
                  "sensor", "upgrades"]

# Initialize dictionary to accumulate aspect ratings for each channel ID
channel_aspect_ratings = {channel_id: {} for channel_id in channel_ids}

# Loop over each channel ID
for channel_id in channel_ids:
    video_id = search_youtube(api_key, channel_id, keyword, max_results=1)

    # if no video about keyword
    if video_id is None:
        print(f"\nNo video ID found for channel: {channel_id}. Unable to proceed.")
    else:
        transcript_text = get_transcript_text(video_id)

        # if the video has no subtitles
        if transcript_text is None:
            print(f"\nNo transcript text available for {keyword} in channel: {channel_id}. Unable to proceed.")
        else:
            data_to_process = wrap_text_into_lines(transcript_text, max_words_per_line=150)

            result = extract_aspects_and_sentiments(inference_source=data_to_process)

            data_list = result  # Replace with your actual data
            aspect_data = extract_data_for_aspects(data_list, target_aspects)

            # Iterate over each aspect and its data
            for aspect, data in aspect_data.items():
                rating = calculate_rating_for_aspect(data)
                if rating is not None:
                    # Accumulate ratings for each channel ID
                    channel_aspect_ratings[channel_id][aspect] = rating

# Calculate the mean ratings for each aspect across all channel IDs
mean_ratings = {}
for aspect in target_aspects:
    aspect_ratings = [channel_aspect_ratings[channel_id].get(aspect, 0) for channel_id in channel_ids if channel_aspect_ratings[channel_id].get(aspect, 0) != 0]
    if aspect_ratings:  # Check if there are non-zero ratings for this aspect
        mean_ratings[aspect] = sum(aspect_ratings) / len(aspect_ratings)

# Plotting Bar Chart for Mean Aspect Ratings
plot_rating_bar_chart(mean_ratings, keyword)
```
